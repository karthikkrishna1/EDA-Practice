```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("IPL_matches.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.drop("umpire3", axis = 1, inplace = True)
```

```python
df.info()
```

```python
len(df.team1.unique())
```

```python
df.team2.value_counts()
```

```python
df.team1.value_counts()
```

```python
df.team1 = np.where((df.team1 == "Rising Pune Supergiant"),"Rising Pune Supergiants",df.team1)
df.team2 = np.where((df.team2 == "Rising Pune Supergiant"),"Rising Pune Supergiants",df.team2)
df.winner = np.where((df.winner == "Rising Pune Supergiant"),"Rising Pune Supergiants",df.winner)
```

```python
df.team1.value_counts()
```

```python
df.team2.value_counts()
```

```python
len(df.winner.unique())
```

```python
len(df.venue.unique())
```

```python
df["date"] = pd.to_datetime(df["date"], format = "%d-%m-%Y")
```

```python
df.sort_values(by = "date")
```

```python
# MI won most matches compared to other teams.
df["winner"].value_counts().plot(kind = "bar");
```

```python
# Ratio of matches won/matches played. 
((df["winner"].value_counts())/((df["team1"].value_counts()) + (df["team2"].value_counts()))).plot(kind = "bar");
# CSK has highest ratio, followed by MI and SRH.
```

```python
df.loc[df["win_by_runs"] != 0].groupby(by = "venue")["win_by_runs"].mean().plot(kind = "bar")
```

```python
df.loc[df["win_by_runs"] != 0].groupby(by = "city")["win_by_runs"].mean().plot(kind = "bar")
```

```python
df.loc[df["win_by_wickets"] != 0].groupby(by = "city")["win_by_wickets"].mean().plot(kind = "bar")
```

```python
df.loc[df["win_by_wickets"] != 0].groupby(by = "venue")["win_by_wickets"].mean().plot(kind = "bar")
```

```python
# It can be seen that teams that won batting second have won with 6 wickets to spare on average in any venue.
# However teams that won batting first have won with varying number of runs to spare.
len(df.loc[df.win_by_wickets != 0]), len(df.loc[df.win_by_runs != 0])
# More teams have won batting 2nd than batting first.
```

```python
(df.team1.value_counts() + df.team2.value_counts()).plot(kind = "bar")
```

```python
numerical_df = df.select_dtypes("number")
sns.boxplot(data = numerical_df, orient = "h")
```

```python
fig, ax = plt.subplots(figsize = (30,10))
sns.countplot(x = "team1",data = df,order = df["team1"].value_counts().index ,ax = ax )
```

```python
fig, ax = plt.subplots(figsize = (30,10))
sns.countplot(x = "winner",order = df["winner"].value_counts().index,data = df, ax = ax )
```
