```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv("zomato.csv", encoding='latin-1')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
 df.isnull().sum()
```

```python
df[df.Cuisines.isnull()].T
```

```python
df[(df.Currency == "Dollar($)") & (df["Has Online delivery"] == 'No') &  (df["Has Table booking"] == 'No') & ((df["Price range"] == 3) | (df["Price range"] == 1))].Cuisines.value_counts()
```

```python
df_country = pd.read_excel("Country-Code.xlsx")
df_country
```

```python
df_combined = df.merge(df_country, on = "Country Code", how = 'left')
```

```python
df_combined
```

```python
fig, ax = plt.subplots()
sns.countplot(x  = "Country", data = df_combined, ax = ax);
plt.xticks(rotation = 45);
```

```python
plt.pie(x = df_combined.Country.value_counts().values[:3], labels =  df_combined.Country.value_counts().index[:3], autopct = "%1.2f%%" );
```

```python
df_combined.columns
```

```python
ratings = df_combined.groupby(["Aggregate rating", "Rating color", "Rating text"]).size().reset_index().rename(columns = {0 : 'Rating count'})
```

```python
ratings
```

## Observation
1. Rating between 4.5 - 4.9 --> Excellent
2. Rating between 4.0 - 4.4 --> Very Good
3. Rating between 3.5 - 3.9 --> Good
4. Rating between 2.5 - 3.4 --> Average
5. Rating between 2.0 - 2.4 --> Poor

```python

matplotlib.rcParams["figure.figsize"] = (12, 16)
sns.barplot(x = "Aggregate rating", y = "Rating count" , data = ratings, hue = "Rating color", palette = ["blue", 'red', 'orange', 'yellow', 'green', 'green']);
plt.xticks(rotation = 90);
```

## Observation
1. Not Rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4

```python
sns.countplot("Rating color", data = ratings,  palette = ["blue", 'red', 'orange', 'yellow', 'green', 'green']);
```
